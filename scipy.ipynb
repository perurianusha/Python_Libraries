```python
#Optimization: Finding the minimum of a function.

import numpy as np
from scipy.optimize import minimize

# Define the function
def func(x):
    return x**2 + 2*x + 1

# Perform the optimization
result = minimize(func, x0=0)  # x0 is the initial guess

print("Optimal value:", result.fun)
print("Optimal x:", result.x)
```

```python
#Integration

from scipy.integrate import quad
import numpy as np

# Define the function
def func(x):
    return np.sin(x)

# Perform the integration
integral, error = quad(func, 0, np.pi)

print("Integral:", integral)
print("Estimated error:", error)
```

```python
#Interpolation
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Define the data points
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])

# Create the interpolation function
f = interp1d(x, y, kind='cubic')

# Define new points to evaluate the interpolation
x_new = np.linspace(0, 5, 100)
y_new = f(x_new)

# Plot the results
plt.plot(x, y, 'o', label='data points')
plt.plot(x_new, y_new, '-', label='cubic interpolation')
plt.legend()
plt.show()
```

```python
#Linear Algebra
import numpy as np
from scipy.linalg import solve

# Define the system of equations
A = np.array([[3, 2], [1, 2]])
b = np.array([5, 5])

# Solve the system
x = solve(A, b)

print("Solution:", x)
```

```python
import numpy as np
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt

# Create a sample signal with noise
np.random.seed(0)
t = np.linspace(0, 1.0, 200)
signal = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(200)

# Design a Butterworth low-pass filter
b, a = butter(4, 0.1)

# Apply the filter
filtered_signal = filtfilt(b, a, signal)

# Plot the results
plt.plot(t, signal, label='Noisy signal')
plt.plot(t, filtered_signal, label='Filtered signal')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt

# Create a sample signal
t = np.linspace(0, 1, 400)
signal = np.sin(50.0 * 2.0 * np.pi * t) + 0.5 * np.sin(80.0 * 2.0 * np.pi * t)

# Compute the Fourier transform
signal_fft = fft(signal)
frequencies = fftfreq(len(t), d=t[1] - t[0])

# Plot the results
plt.plot(frequencies, np.abs(signal_fft))
plt.title("Fourier Transform")
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.show()
```

```python
import numpy as np
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import spsolve

# Create a sparse matrix
A = csr_matrix([[3, 0, 0], [0, 4, 0], [0, 0, 5]])
b = np.array([1, 2, 3])

# Solve the system
x = spsolve(A, b)

print("Solution:", x)
```

```python
import numpy as np
from scipy.stats import ttest_ind

# Generate sample data
np.random.seed(0)
data1 = np.random.normal(0, 1, 100)
data2 = np.random.normal(0.5, 1, 100)

# Perform a t-test
t_stat, p_value = ttest_ind(data1, data2)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

# Define points
points = np.array([[0, 0], [1, 1], [2, 2]])

# Compute pairwise distances
distances = pdist(points, metric='euclidean')

# Convert to a square form matrix
distance_matrix = squareform(distances)

print("Pairwise distances:\n", distance_matrix)
```
